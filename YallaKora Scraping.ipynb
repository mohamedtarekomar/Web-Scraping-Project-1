# In this notebook i will bring data from "Yallakora.com" website..the data will be about matches information from 8/18/2023 to 8/27/2023 (10 pages) and the data will be : 

## 1) Date of the match 
## 2) Champion title
## 3) Match(teams)
## 4) Score
## 5) Time of the match 
## 6) Status of the match 
## 7) Channel that will Broadcast the match
## 8) Round


```python
# Importing "BeautifulSoup" and "requests" libraries to get the data from the website

from bs4 import BeautifulSoup
import requests
```

```python
# Make an empty lists to append in later

Channel = []
Status = []
Time = []
Score = []
Matches = []
championTitle = []
Round = []
Date = []
```

```python
# Loop in range 10 days 
for k in range (18,28) :
    # Requests the url title 
    response = requests.get("https://www.yallakora.com/match-center/?date=8/{k}/2023#")
    html_content = response.content
    # Change the content to "html.parser" to be easy to read
    soup = BeautifulSoup(html_content, "html.parser")
    
    # The class "matchCard" includes all the data i need
    matchCard = soup.find_all("div", class_ = "matchCard")

    # The length of the "matchCard" is the number of the champion titles
    for j in range(len(matchCard)) :
        
        # This range means a number of the matches in each champion
        for i in range(len(matchCard[j].contents[3].find_all("li"))) :
            
            TeamA = (matchCard[j].contents[3].find_all("li")[i].find("div",class_ = "teams teamA").text.strip())
            TeamB = (matchCard[j].contents[3].find_all("li")[i].find("div",class_ = "teams teamB").text.strip())
            Matches.append(f'{TeamA} - {TeamB} ')
            Channel.append(matchCard[j].contents[3].find_all("li")[i].find("div", class_="channel icon-channel").text.strip() if matchCard[j].contents[3].find_all("li")[i].find("div", class_="channel icon-channel") else "N/A")
            Status.append(matchCard[j].contents[3].find_all("li")[i].find("div",class_ = "matchStatus").text.strip())
            Time.append(matchCard[j].contents[3].find_all("div","MResult")[i].find_all("span",class_ = "time")[0].text.strip())
            resultA = (matchCard[j].contents[3].find_all("div","MResult")[i].find_all("span",class_ = "score")[0].text.strip())
            resultB = (matchCard[j].contents[3].find_all("div","MResult")[i].find_all("span",class_ = "score")[1].text.strip())
            Score.append(f'{resultA} - {resultB} ')  
            championTitle.append(matchCard[j].contents[1].find("h2").text.strip())
            Round.append(matchCard[j].contents[3].find_all("div", class_="date")[i].text.strip())
            Date.append(f"8/{k}/2023")

print(Channel)
print(Status)
print(Time)
print(Score)
print(Matches)
print(championTitle)
print(Round)
print(Date)
```

```python
# Import pandas library to convert the data from lists into dataframe

import pandas as pd

# Make the data dic to convert it into dataframe 
data = {
    "Date" : Date,
    "Matches" : Matches,
    "Champion Title" : championTitle,
    "Time" : Time,
    "Score" : Score, 
    "Channel" : Channel,
    "Status" : Status,
    "Round" : Round,
    
}
```

```python
data
```

```python
# Converting the data to be dataframe

df = pd.DataFrame(data)
df
```

```python
# Save the df to my device as a csv file

df.to_csv("Yallakora Scraping.csv", encoding='utf-8-sig', index=False)
```
